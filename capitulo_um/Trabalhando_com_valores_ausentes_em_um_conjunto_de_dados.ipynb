```python
import pandas as pd
import numpy as np

def cat_frequente(dataframe, coluna):
    moda = dataframe[coluna].mode()[0]
    valores_trocados = dataframe[coluna].fillna(moda)
    return valores_trocados

dicionario = {
    "sobrenome": ["Silva", "Cavalcanti", "Oliveira", "Santos", "Souza", "Lima", "Pereira", "Rodrigues", "Carvalho", "Alves"],
    "idade": [71, 35, 61, 73, 28, 41, 35, 70, 46, 68],
    "genero": ["Masculino", np.nan, "Feminino", "Masculino", "Feminino", np.nan, "Feminino", np.nan, "Masculino", "Feminino"]
}

dataframe = pd.DataFrame(dicionario)
dataframe["genero_alterado"] = cat_frequente(dataframe, "genero")

dataframe.head()
```

```python
dataframe.to_csv("dataframe.csv",index=False)
```

## Somando valores nulos do nosso DataFrame

```python
def soma_desconsiderando_nulos(vetor):
    soma = 0
    for valor in vetor:
        if valor is not None:
            soma += valor
    return soma
```

```python
soma_desconsiderando_nulos(dataframe["idade"])/5
```

```python
def troca_media(dataframe,coluna):
    media = dataframe[coluna].mean()
    valores_trocados = dataframe[coluna].fillna(media)

    return valores_trocados
```

```python
vetor = [10,np.nan,1,9,np.nan,20,np.nan,14,2,20]
dataframe = pd.DataFrame({"valores":vetor})
dataframe["valores_trocados"] = troca_media(dataframe, "valores")
```

```python
def troca_mediana(dataframe,coluna):
    mediana = dataframe[coluna].median()
    valores_trocados = dataframe[coluna].fillna(mediana)

    return valores_trocados
```

```python
vetor = [10,np.nan,1,9,np.nan,20,np.nan,14,2,20]

dicionario2 = {"valores":vetor}
dataframe = pd.DataFrame(dicionario2)
dataframe["valores_trocados"] = troca_mediana(dataframe,"valores")
```
