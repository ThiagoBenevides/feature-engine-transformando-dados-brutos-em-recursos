```python
entrada =pd.date_range("2020-05-01",periods=31,freq="H")
saida = pd.date_range("2022-05-01",periods=31, freq="M")

dataframe =pd.DataFrame({"entrada":entrada,"saida":saida})
```

```python
dataframe.head()
```

```python
dataframe["diferenca"] = (dataframe["saida"]-dataframe["entrada"]).dt.days
```

```python
dataframe.head()
```

```python
dataframe["diferenca_meses"] = (dataframe["saida"]-dataframe["entrada"])//np.timedelta64(1,"M")
```

```python
dataframe.head()
```

```python
dataframe["diferencao_segundos"] = (dataframe["saida"]-dataframe["entrada"])//np.timedelta64(1,"s")
dataframe["diferencao_minutos"]= (dataframe["saida"]-dataframe["entrada"])//np.timedelta64(1,"m")
```

```python
tempo_1 = pd.Series(pd.date_range(start="2022-01-01 09:00",freq="H", periods=5, tz= "America/Sao_Paulo"))

tempo_2 = pd.Series(pd.date_range(start="2022-01-01 09:00",freq="H", periods=5,tz="US/Central"))

aux_1 = pd.concat([tempo_1,tempo_2],axis=0)

aux_2 = pd.concat([pd.Series(
    pd.date_range(start="2022-01-01 12:00",freq="H",periods=5,tz="Europe/Berlin")),
    pd.Series(
        pd.date_range(
            start="2022-03-01 13:00",freq="H", periods=5,
            tz="US/Central"))
    ],axis=0)

timezone= pd.DataFrame(index=range(0,10))
timezone["data_1"]= aux_1.values
timezone["data_2"]= aux_2.values
```

```python
timezone.head(10)
```

```python
timezone["tempo1_utc"]= pd.to_datetime(timezone["data_1"],utc=True)
timezone["tempo2_utc"] = pd.to_datetime(timezone["data_2"], utc=True)
```
